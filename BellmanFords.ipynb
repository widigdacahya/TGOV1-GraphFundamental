```python
def bellman_ford(nodes, edges, source_index=1):
    path_length = {v: float('inf') for v in nodes}
    path_length[source_index] = 0

    paths = {v: [] for v in nodes}
    paths[source_index] = [source_index]

    for _ in range(len(nodes) - 1):
        for(u, v), w_uv in edges.items():
            if path_length[u] + w_uv < path_length[v]:
                path_length[v] = path_length[u] + w_uv
                paths[v] = paths[u] + [v]


    for (u,v), w_uv in edges.items():
        if path_length[u] + w_uv < path_length[v]:
            print("Negative cycle detected")

    return path_length, paths
```
